```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('noise-long-data.txt')
```

```python

y = data[2:].astype(float).values.flatten()
l = len(y)
```

```python
t = np.linspace(0.,float(l-1)/44100.,num=l)
```

```python
ps = np.abs(np.fft.fft(y))**2
time_step = 1/44100.
freqs = np.fft.fftfreq(y.size, time_step)
idx = np.argsort(freqs)
```

```python
#print(idx)
plt.plot(freqs[idx[int(l/2-1):int(9*l/16)]],ps[idx[int(l/2)-1:int(9*l/16)]])
```
