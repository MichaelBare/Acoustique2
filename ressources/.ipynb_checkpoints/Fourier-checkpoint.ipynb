```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('noise-data.txt')
```

```python
t = np.linspace(0.,536./44100.,num=537)
```

```python
y = data[2:].astype(float).values.flatten()
```

```python
ps = np.abs(np.fft.fft(y))**2
time_step = 1/44100.
freqs = np.fft.fftfreq(y.size, time_step)
idx = np.argsort(freqs)
```

```python
#print(idx)
plt.plot(freqs[idx[268:268+50]],ps[idx[268:268+50]])
```
